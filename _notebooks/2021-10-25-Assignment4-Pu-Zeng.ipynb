## Data Preparation

### Combine Selected Datasets regarding Race Variable

This report is tring to illustrate these following study aims:   
1. How does the number of Doctor degree receivers in each types of study change when stratified by gender or race?  
2. How does the number of Dctor degree receivers in each types of employment change when stratified by race? 
2. How does the number of Dctor degree receivers in each types of debt levels change when stratified by race?   

For that, a few data tables were selected to discuss the distribution of race and gender in the variables of:
1. Major field of study
2. Employement types
3. Debt levels

Prior to create the dashboard visualization, there are some data processing need to be done.

First, we combined a few datatables regarding the variable, race.

#### Race vs Major Field of Study vs Employment Types vs Debt Level

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
debt_race_url = 'https://raw.githubusercontent.com/puzeng/BIOSTAT823_FALL21_Pu/main/data/raw/Debt_race_2017.csv'
```

```python
debt_race_df = pd.read_csv(debt_race_url)
debt_race_df.head()
```

```python
debt_race_df  = debt_race_df.melt(id_vars = 'Debt level', 
                                  var_name = 'Ethnicity status', 
                                  value_name = 'Count')
debt_race_df.head()
```

```python
employment_race_url = 'https://raw.githubusercontent.com/puzeng/BIOSTAT823_FALL21_Pu/main/data/raw/Employment_race_2017.csv'
employment_race_df = pd.read_csv(employment_race_url)
employment_race_df.head()
```

```python
employment_race_df = employment_race_df.melt(id_vars = 'Employment Types', 
                                             var_name = 'Ethnicity status')
employment_race_df.head()
```

```python
race_study_field_url = 'https://raw.githubusercontent.com/puzeng/BIOSTAT823_FALL21_Pu/main/data/raw/Race_Fine_Field_Study.csv'
race_study_field_df = pd.read_csv(race_study_field_url)
race_study_field_df.head()
```

```python
race_study_field_df = race_study_field_df.melt(id_vars = 'Field of study', 
                                               var_name = 'Ethnicity status', 
                                               value_name = 'Number of count')
race_study_field_df.head()
```

```python
race_phd_info = (
    race_study_field_df.
    merge(employment_race_df, on = 'Ethnicity status', how = 'left' ).
    merge(debt_race_df, on = 'Ethnicity status', how = 'left' )
)
race_phd_info.head()
```

#### Sex vs Major Field of Study

We then will add the sex column to the below data table.  

```python
sex_field_study_url = 'https://raw.githubusercontent.com/puzeng/BIOSTAT823_FALL21_Pu/main/data/raw/Sex_Fine_Filed_Study.csv'
sex_field_study_df = pd.read_csv(sex_field_study_url)
sex_field_study_df.head()
```

```python
male_index = sex_field_study_df.iloc[:, 0][sex_field_study_df.iloc[:, 0] == 'Male'].index.tolist()
female_index = sex_field_study_df.iloc[:, 0][sex_field_study_df.iloc[:, 0] == 'Female'].index.tolist()
```

```python
ALL = sex_field_study_df.iloc[:male_index[0], :]
MALE = sex_field_study_df.iloc[male_index[0]:female_index[0], :].reset_index(drop = True)
FEMALE = sex_field_study_df.iloc[female_index[0]:, :].reset_index(drop = True)
```

```python
ALL = ALL.assign(Sex = 'All')
ALL = ALL.reindex(ALL.index.drop(0)).reset_index(drop = True)
ALL.head()
```

```python
MALE = MALE.assign(Sex = 'Male')
MALE = MALE.reindex(MALE.index.drop(0)).reset_index(drop = True)
MALE.head()
```

```python
FEMALE = FEMALE.assign(Sex = 'Female')
FEMALE = FEMALE.reindex(FEMALE.index.drop(0)).reset_index(drop = True)
FEMALE.head()
```

```python
sex_field_study_df = pd.concat([ALL, MALE, FEMALE],ignore_index=True)
sex_field_study_df.head()
```

## Dashboard Visualization

You can also look at the race distribution against the variable, major field of study, employment types, and debt levels.  

How to interact with the dashboard is detailedly described in the web app. The script, `Streamlit_App` for building the web app can be retrieved from this GitHub repo [Streamlit Web App](https://github.com/puzeng/BIOSTAT823_Streamlit).
